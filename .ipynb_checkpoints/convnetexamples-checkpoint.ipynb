```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_val, y_val) = mnist.load_data()

x_train = x_train.reshape(-1,28,28,1).astype('float32') / 255.0
x_val = x_val.reshape(-1,28,28,1).astype('float32') / 255.0

y_train = keras.utils.to_categorical(y_train)
y_val = keras.utils.to_categorical(y_val)
```

```python
from tensorflow.keras.layers import Conv2D, SeparableConv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization, Activation
from tensorflow.keras import regularizers

model = keras.models.Sequential()

model.add(Conv2D(filters=6, kernel_size=5, strides=1, use_bias=True, padding='valid', input_shape=(28, 28, 1,), activation='relu'))
model.add(BatchNormalization(axis=-1))
model.add(MaxPooling2D(pool_size=(2,2), padding='valid'))

model.add(Conv2D(filters=6, kernel_size=5, strides=1, use_bias=False, padding='valid', activation='relu'))
model.add(BatchNormalization(axis=-1))
model.add(MaxPooling2D(pool_size=(2,2), padding='valid'))

model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(100))
model.add(Activation('relu'))

model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])

model.summary()
```

```python
hist = model.fit(x_train, y_train, epochs=20, batch_size=64, validation_data=(x_val,y_val))
```

```python
acc = hist.history['acc']
val_acc = hist.history['val_acc']
loss = hist.history['loss']
val_loss = hist.history['val_loss']

plt.plot(acc, 'bo', label='Training acc')
plt.plot(val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.show()

plt.show()
```

```python
from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

(x_train, y_train), (x_test, y_test) = mnist.load_data()

scaler = StandardScaler().fit(x_train.reshape(x_train.shape[0],-1))
scaled_x_train = scaler.transform(x_train.reshape(x_train.shape[0],-1))
scaled_x_test = scaler.transform(x_test.reshape(x_test.shape[0],-1))
```

```python
target_dim = 30
encoder = keras.models.Sequential()
encoder.add(keras.layers.Dense(750, input_shape=(scaled_x_train.shape[1],), activation='relu'))
encoder.add(keras.layers.Dense(500, activation='relu'))
encoder.add(keras.layers.Dense(250, activation='relu'))
encoder.add(keras.layers.Dense(target_dim, activation='relu'))
encoder.summary()

decoder = keras.models.Sequential()
decoder.add(keras.layers.Dense(250, input_shape=(target_dim,), activation='relu'))
decoder.add(keras.layers.Dense(500, activation='relu'))
decoder.add(keras.layers.Dense(750, activation='relu'))
decoder.add(keras.layers.Dense(scaled_x_train.shape[1], activation='relu'))
decoder.summary()

encdec = keras.models.Sequential()
encdec.add(encoder)
encdec.add(decoder)
encdec.compile(optimizer='Adam', loss='mse')
encdec.summary()
```

```python
hist = encdec.fit(scaled_x_train, scaled_x_train, epochs=15, batch_size=64)
```

```python
x_test_ed = encdec.predict(scaled_x_test)

fig, ax = plt.subplots(5,4)
ax = ax.flatten()

for i in range(10):
    im_idx = np.argwhere(y_test == i)
    pick = np.random.choice(y_test)
    plottable_image = np.reshape(x_test[im_idx][pick], (28, 28))
    another_image = np.reshape(scaler.inverse_transform(x_test_ed[im_idx][pick]), (28, 28))
    ax[i*2].imshow(plottable_image, cmap='gray_r')
    ax[i*2+1].imshow(another_image, cmap='gray_r')
```
